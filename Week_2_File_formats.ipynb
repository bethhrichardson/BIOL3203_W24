Now, let's look at some sequence data.

````python
#A DNA sequence is just a string of characters that can be easily manipulated by a computer. We can work with strings in base Python.

x = "ATGGCGGGAAAATGA"

print(x)

print(x[5])

#Now try printing the first T

#Now try printing the last G
````

````python
##We can also work with DNA sequence strings in BioPython.
##This may look like more work now, but when you're working with real sequence data it will be MUCH easier.

from Bio.Seq import Seq
x = Seq("ATGGCGGGAAAATGA")
Seq('ATGGCGGGAAAATGA')
x[5]
````

````python
##Let's try transcribing a sequence
from Bio.Seq import Seq
x = Seq("ATGGCGGGAAAATGA")
print(x.transcribe())

##Let's try translating a sequence
print(x.translate())

##Let's try measuring the GC content of a sequence
from Bio.SeqUtils import gc_fraction
print(f"GC content of {x} : {gc_fraction(x):.2f}")
````

So as well as just defining a DNA sequence as a string, using BioPython allows us to carry out biological operations like transcribing and translating DNA. You can also assign these items to different variables - instead of printing the transcription of x, you can save it as x.RNA.



